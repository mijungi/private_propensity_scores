# Effect of $\alpha$ on Average Treatment Effect

```python
import copy

import numpy as np
import matplotlib.pyplot as plt
import torch

from agm import calibrateAnalyticGaussianMechanism

%matplotlib inline

# set random seed
np.random.seed(1)
torch.manual_seed(1)
```

```python
# no.experiments, no. bootstrap draws, no. samples, dim, X
ne = 100
k = 10
ns = 1200
dim = 50

# draw ne separate ns samples
X_std = 1
X_dist = torch.distributions.normal.Normal(torch.tensor([0.0], dtype=torch.float64), torch.tensor([X_std], dtype=torch.float64))
X = [X_dist.sample((ns, dim)).squeeze() for i in range(ne)]

# restrict X to ||x||_2 \leq 1 to fit assumption for each experiment
X = torch.stack([X[i] / X[i].norm(dim=1).max() for i in range(ne)])
```

```python
# no. of points used to fit log reg
nf = 1000
# sampling rate
sam_rate = 1 / k

# privacy parameters
epses = [8.806244425107934, 4.576545956409196, 3.076600043713613, 1.3471128488311916, 0.9194619944192267]
alphas = [1.01, 1.5, 2, 4, 6]

# regularisation coefficient
reg_co = 0.1
```

```python
# vectors for generating Y, differentiate between Y_0 and Y_1 with bias
# Y = beta^T X + 0.1 Z
# Y_1 = Y + bias, Y_0 = Y
beta_std = 1
beta_dist = torch.distributions.normal.Normal(
    torch.tensor([0.0], dtype=torch.float64), torch.tensor([beta_std], dtype=torch.float64))
beta = beta_dist.sample((dim, 1)).reshape(dim, 1)
bias = 0.25

# generate Y
rand_mult = 0.1
Y = torch.einsum('kl,ijk->ij',beta,X) + rand_mult * torch.randn(ne, ns, dtype=torch.float64)
```

```python
# vectors for generating T
# T = exp(-T_w^T X + b)
T_std = 3
T_dist = torch.distributions.normal.Normal(
    torch.tensor([0.0], dtype=torch.float64), torch.tensor([T_std], dtype=torch.float64))
T_w = T_dist.sample((dim, 1)).reshape(dim, 1)
T_b = 0

# generate T \in {0, 1}
prob_vec = torch.sigmoid(torch.einsum('kl,ijk->ij', T_w, X) + T_b)
T = torch.bernoulli(prob_vec)
```

```python
class Log_Reg(torch.nn.Module):
    '''
    Logistic Regression
    '''
    def __init__(self, D_in, D_out):
        super(Log_Reg, self).__init__()
        self.linear = torch.nn.Linear(D_in, D_out, bias=False)
        
    def forward(self, x):
        y_pred = torch.sigmoid(self.linear(x))
        return y_pred
```

```python
def IPTW_PPS_MC(X, T, prob_vec, Y, bias, alphas, reg_co, k, sam_rate, nf):
    '''
    estimate treatment effect with inverse propensity weighting using private propensity scores
    '''
    # get # experiments, # samples, # dimensions
    ne, ns, dim = X.shape
    
    # sgd step size
    step_size = 0.01
    
    ################
    # process data #
    ################
    
    # get Y0 and Y1
    Y0 = Y * (1 - T)
    Y1 = (Y + bias) * T
    print(Y0.mean(), Y1.mean())
    
    # split data
    # get splits
    fit_split = nf
    est_split = ns - nf

    # get # bootstrap draws
    nb = int(sam_rate * nf)
    
    # permute indices
    perm = torch.stack([torch.randperm(ns) for i in range(ne)])

    # create splits
    s0 = perm[:, :fit_split]
    s1 = perm[:, fit_split:]

    # create auxiliary indices
    idx = torch.arange(ne)[:, None]
    
    # create sampling indices
    k_perm = torch.arange(nf).reshape(k, nb, 1)

    # split X into fit, estimate splits
    X_s0 = X[idx, s0]
    X_s1 = X[idx, s1]

    # expand dim of T to allow multiplication with X
    T_ex_dim = T.reshape(ne, ns, 1)

    # split X0 and X1 into fit, estimate splits
    X0_s1 = (X * (1- T_ex_dim))[idx, s1]
    X1_s1 = (X * T_ex_dim)[idx, s1]

    # precompute ||X0_sl||_2^2 and ||X1_sl||_2^2
    X0_s1_2 = (X0_s1.norm(dim=2) ** 2).reshape(ne, 1, est_split)
    X1_s1_2 = (X1_s1.norm(dim=2) ** 2).reshape(ne, 1, est_split)

    # add x_i to x_j for X0_plus_X0_s1 and X1_plus_X1_s1
    X0_plus_X0_s1 = X0_s1.reshape(ne, est_split, 1, dim) + X0_s1.reshape(ne, 1, est_split, dim)
    X1_plus_X1_s1 = X1_s1.reshape(ne, est_split, 1, dim) + X1_s1.reshape(ne, 1, est_split, dim)

    # get norms squared for X0_plus_X0_s1 and X1_plus_X1_sl
    X0_plus_X0_sl_norm_2 = (X0_plus_X0_s1.norm(dim=-1) ** 2).reshape(ne, 1, est_split, est_split) 
    X1_plus_X1_sl_norm_2 = (X1_plus_X1_s1.norm(dim=-1) ** 2).reshape(ne, 1, est_split, est_split)

    # split T into fit, estimate splits
    T_s0 = T[idx, s0]
    T_s1 = T[idx, s1]

    # split Y0 and Y1 into fit, estimate splits 
    Y0_s0 = Y0[idx, s0] 
    Y1_s0 = Y1[idx, s0] 

    Y0_s1 = Y0[idx, s1] 
    Y1_s1 = Y1[idx, s1] 

    # mult y_i to y_j for Y0_times_Y0_s1 and Y1_times_Y1_s1
    Y0_times_Y0_s1 = Y0_s1.reshape(ne, 1, est_split, 1) * Y0_s1.reshape(ne, 1, 1, est_split)
    Y1_times_Y1_s1 = Y1_s1.reshape(ne, 1, est_split, 1) * Y1_s1.reshape(ne, 1, 1, est_split)
                            
    # reshape estimate splits for later
    Y0_s1 = Y0_s1.reshape(ne, 1, est_split)
    Y1_s1 = Y1_s1.reshape(ne, 1, est_split)
    

    ##############
    # fit models #
    ##############

    # instantiate ne different models and set model parameters to float64
    models = [[Log_Reg(dim, 1).double() for j in range(k)] for i in range(ne)]

    # define loss (binary cross entropy)
    loss = torch.nn.BCELoss()

    # define optimisers
    optimisers = [[torch.optim.SGD(models[i][j].parameters(), 
                                   lr=step_size, weight_decay=reg_co) for j in range(k)] for i in range(ne)]
    
    # train model
    for t in range(1000):
        preds = [[models[i][j](X_s0[i, k_perm[j, :].squeeze()]).squeeze() for j in range(k)] for i in range(ne)]
        losses = [[loss(preds[i][j], T_s0[i, k_perm[j, :].squeeze()]) for j in range(k)] for i in range(ne)]
        [[optimisers[i][j].zero_grad for j in range(k)] for i in range(ne)]
        [[losses[i][j].backward() for j in range(k)] for i in range(ne)]
        [[optimisers[i][j].step() for j in range(k)] for i in range(ne)]
                        
    #############################    
    # estimate treatment effect #
    #############################
    
    # initialise pi_hat dictionaries 
    pi_hats = {} 
    pi_hats_analytic = {}
        
    # get estimated propensity scores
    pi_hats[0] = torch.stack([torch.stack([models[i][j](X_s1[i]).squeeze() for j in range(k)]) for i in range(ne)])
    
    # perturb model and get relevant quantities
    for eps in alphas:
        # define sigma
        s_a = 2. / (nb * reg_co)

        # gaussian mechanism
        # set sigma
        
        # vanilla
        # sigma = np.sqrt(2 * np.log(1.25 / delta) + 1e-6) *  (s_a  / eps)
        # sigma_2 = sigma ** 2

        # moments accountant
        sigma = s_a * eps
        sigma_2 = sigma ** 2
        
        # # analytic gaussian mechanism
        # sigma = calibrateAnalyticGaussianMechanism(eps, delta, s_a)
        # sigma_2 = sigma ** 2

        # define noise distribution
        noise_dist = torch.distributions.normal.Normal(torch.tensor([0.0], dtype=torch.float64), torch.tensor([sigma], dtype=torch.float64))

        # draw noise vectors
        noise_vecs = noise_dist.sample((ne, k, dim)).squeeze()

        # create temp models 
        models_ = copy.deepcopy(models)

        # \hat{\w}^\top Xs
        w_T_X0_s1 = []
        w_T_X0_plus_X0_s1 = []
        w_T_X1_s1 = []
        w_T_X1_plus_X1_s1 = []

        # initialise list for privatised estimated propensity scores
        pi_hats[eps] = []
        
        # perturb weights with noise vectors
        for i in range(ne):
            for j in range(k):
                w_T_X0_s1.append(torch.einsum('ij,kj-> i', X0_s1[i], models[i][j].linear.weight))
                w_T_X0_plus_X0_s1.append(torch.einsum('ijk,lk-> ij', X0_plus_X0_s1[i], models[i][j].linear.weight))
                w_T_X1_s1.append(torch.einsum('ij,kj-> i', X1_s1[i], models[i][j].linear.weight))
                w_T_X1_plus_X1_s1.append(torch.einsum('ijk,lk-> ij', X1_plus_X1_s1[i], models[i][j].linear.weight))
                # perturb i, j weights
                model_temp = models_[i][j]
                model_temp.linear.weight.data.add_(noise_vecs[i, j, :])
                pi_hats[eps].append(model_temp(X_s1[i]).squeeze())
                
        # reshape stacked privatised estimated propensity scores as ne * nd
        pi_hats[eps] = torch.stack(pi_hats[eps]).reshape(ne, k, est_split)

        # precompute sigma^2 ||x||_2^2 and \w^\top Xs
        pi_hats_analytic[eps] = {'sigma_2_X0_s1_2': sigma_2 * X0_s1_2}
        pi_hats_analytic[eps]['sigma_2_X1_s1_2'] = sigma_2 * X1_s1_2
        pi_hats_analytic[eps]['sigma_2_X0_plus_X0_s1'] = sigma_2 * X0_plus_X0_sl_norm_2
        pi_hats_analytic[eps]['sigma_2_X1_plus_X1_s1'] = sigma_2 * X1_plus_X1_sl_norm_2
        pi_hats_analytic[eps]['w_T_X0_s1'] = torch.stack(w_T_X0_s1).reshape(ne, k, est_split)
        pi_hats_analytic[eps]['w_T_X0_plus_X0_s1'] = torch.stack(w_T_X0_plus_X0_s1).reshape(ne, k, est_split, est_split)
        pi_hats_analytic[eps]['w_T_X1_s1'] = torch.stack(w_T_X1_s1).reshape(ne, k, est_split)
        pi_hats_analytic[eps]['w_T_X1_plus_X1_s1'] = torch.stack(w_T_X1_plus_X1_s1).reshape(ne, k, est_split, est_split)
    
    # get treatment effects
    # true
    te = {}
    # empirical means and stds of means of ERM + noise 
    te_hats = {'means': [], 'stds': []}
    # analytic means and stds of ERM + noise
    te_hats_analytic = {'means': [], 'stds': []}

    # estimate true treatment effect
    te_ = torch.mean(Y1_s1.squeeze() / prob_vec[idx, s1] - Y0_s1.squeeze() / (1 - prob_vec[idx, s1]), 1)
    te['mean'] = te_.mean().detach().numpy()
    te['std'] = te_.std().detach().numpy()

    for key in pi_hats.keys():
        # empirical estimates
        # reduce_mean from (ne, k, est_split) tensor to (ne * nd, 1) matrix
        te_hats_ = torch.mean(Y1_s1 / pi_hats[key] - Y0_s1 / (1 - pi_hats[key]), 2)
        te_hats['means'].append([te_hats_[i].mean().detach().numpy() for i in range(ne)])
        te_hats['stds'].append([te_hats_[i].std().detach().numpy() for i in range(ne)])
        if key != 0:
            # analytic estimates
            # expectation and variance of mu_0
            rand_mu_0 = Y0_s1 * torch.exp(pi_hats_analytic[key]['w_T_X0_s1'] + pi_hats_analytic[key]['sigma_2_X0_s1_2']/2)
            E_mu_0 = torch.mean(Y0_s1 + rand_mu_0, [2])
            mu_X0_s1_2 = Y0_times_Y0_s1 * torch.exp(pi_hats_analytic[key]['w_T_X0_plus_X0_s1'] + pi_hats_analytic[key]['sigma_2_X0_plus_X0_s1']/2)
            mu_X0_s1_mu_X0_s1 = rand_mu_0.reshape(ne, k, 1, est_split) * rand_mu_0.reshape(ne, k, est_split, 1)
            var_mu_0 = torch.mean(mu_X0_s1_2 - mu_X0_s1_mu_X0_s1, [2, 3])
            # expectation and variance of mu_1
            rand_mu_1 = Y1_s1 * torch.exp(- pi_hats_analytic[key]['w_T_X1_s1'] + pi_hats_analytic[key]['sigma_2_X1_s1_2']/2)
            E_mu_1 = torch.mean(Y1_s1 + rand_mu_1, [2])
            mu_X1_s1_2 = Y1_times_Y1_s1 * torch.exp(- pi_hats_analytic[key]['w_T_X1_plus_X1_s1'] + pi_hats_analytic[key]['sigma_2_X1_plus_X1_s1']/2)
            mu_X1_s1_mu_X1_s1 = rand_mu_1.reshape(ne, k, 1, est_split) * rand_mu_1.reshape(ne, k, est_split, 1)
            var_mu_1 = torch.mean(mu_X1_s1_2 - mu_X1_s1_mu_X1_s1, [2, 3])
            # expectation and variance of te_hats
            te_hats_analytic_mu = E_mu_1 - E_mu_0
            te_hats_analytic_std = torch.sqrt(var_mu_1 + var_mu_0)
            te_hats_analytic['means'].append([te_hats_analytic_mu[i].mean().detach().numpy() for i in range(ne)])
            te_hats_analytic['stds'].append([te_hats_analytic_std[i].mean().detach().numpy() for i in range(ne)])

    te_hats['means'] = np.array(te_hats['means'])
    te_hats['stds'] = np.array(te_hats['stds'])
    te_hats_analytic['means'] = np.array(te_hats_analytic['means'])
    te_hats_analytic['stds'] = np.array(te_hats_analytic['stds'])
 
    return te, te_hats, te_hats_analytic
```

```python
def plot_hist(figname, bias, te_hats, var, ne):
    '''
    plot histogram of empirical probabilities flipped signs 
    '''
    # deal with nans by setting them to be -bias
    for i in range(len(te_hats['means'][1:])):
        te_hats['means'][1:][i][np.isnan(te_hats['means'][1:][i])] = -bias     
    
    # compute probabilities
    if bias > 0:
        probs = [sum(i < 0) / ne for i in te_hats['means'][1:]]
    else:
        probs = [sum(i > 0) / ne for i in te_hats['means'][1:]]
        
    str_var = [str(i) for i in var]
    
    fig, ax = plt.subplots(1,1, figsize=(20, 10))
    ax.bar(str_var, probs)

    ax.set_title("Probability of Flipped Sign against $\\alpha$", fontsize=20)
    ax.set_ylabel("Probability of Flipped Sign", fontsize=20)
    ax.set_xlabel("$\\alpha$", fontsize=20)

    fig.tight_layout()
    fig.savefig(figname+'.png',dpi=100)
```

```python
def plot_te(figname, te, te_hats, te_hats_analytic, epses, cutoff, bias):
    '''
    plot the true treatment effect, ERM, private ERM treatment effect
    '''
    fig, ax = plt.subplots(1,1, figsize=(20, 10))

    te_hat = np.mean([te_hats['means'][0]], 1)
    te_hat_std = np.std([te_hats['means'][0]])
    te_hat_z = np.mean(te_hats['means'][1:cutoff+1], 1)
    te_hat_z_std = np.std(te_hats['means'][1:cutoff+1], 1)
    te_hat_mu = np.mean(te_hats_analytic['means'][:cutoff], 1)
    te_hat_mu_std = np.std(te_hats_analytic['means'][:cutoff], 1) + np.mean(te_hats_analytic['stds'][:cutoff], 1)
    
    ax.plot(epses[:cutoff], [te['mean']] * len(epses[:cutoff]) , marker='o', color='magenta', label="Truth")
#     ax.fill_between(epses, te['mean'] + te['std'], te['mean'] - te['std'], facecolor='magenta', alpha=0.25)
    ax.plot(epses[:cutoff], [te_hat] * len(epses[:cutoff]) , marker='o', color='red', label="ERM")
    ax.plot(epses[:cutoff], te_hat_z, marker='x', color='blue', label="Empirical Private ERM")
    ax.plot(epses[:cutoff], te_hat_mu ,marker='d', color='green', label="Analytical Private ERM")
    ax.fill_between(epses[:cutoff], te_hat + te_hat_std, 
                        te_hat - te_hat_std, facecolor='red', alpha=0.25)
    ax.fill_between(epses[:cutoff], te_hat_z + te_hat_z_std, 
                        te_hat_z - te_hat_z_std, facecolor='blue', alpha=0.25)
    ax.fill_between(epses[:cutoff], te_hat_mu + te_hat_mu_std,
                        te_hat_mu - te_hat_mu_std, facecolor='green', alpha=0.25)

    ax.set_title("True, ERM, Empirical Private ERM and Analytical Private ERM of Average Treatment Effect against $\\alpha$", fontsize=20)
    ax.set_ylabel("Average Treatment Effect", fontsize=20)
    ax.set_xlabel("$\\alpha$", fontsize=20)
    if bias > 0:
        ax.legend(fontsize=16, loc=1)
    else:
        ax.legend(fontsize=16, loc=4)

    fig.tight_layout()
    fig.savefig(figname+'.png',dpi=100)
```
